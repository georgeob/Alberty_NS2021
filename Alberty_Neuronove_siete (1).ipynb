```python
import numpy as np #Importovanie knižníc na spracovanie dát a súborov 
import pandas as pd
```

```python
df = pd.read_csv('tesla_april.csv', parse_dates=['datetime'], infer_datetime_format=True) # načítanie prvého csv súboru kotrý som pomenoval df
```

```python
df['date'] = df.datetime.dt.date #Vytvorenie stĺpca date- naplnenie zo stĺpca datetime (dátum)
df['time'] = df.datetime.dt.time #Vytvorenie stĺpca time- naplnenie zo stĺpca datetime (čas)
df #výpis súboru csv
```

```python
df['date'] = pd.to_datetime(df['date']) #Konvertovanie stĺpcu date na typ datetime
```

```python
df1=(df[df["date"]=='2020-03-30'][:160]) #Vytvorenie novej premennej df1 - k nej bolo priradených prvýchh 160 riadkov zo stĺpca date z premennej df, ktoré sa rovnali dátumu 2020-03-30
df1 #výpis premennej df1
```

```python
df2=(df[df["date"]=='2020-04-29'][:160]) #Vytvorenie novej premennej df2 - k nej bolo priradených prvých 160 riadkov zo stĺpca date z premennej df, ktoré sa rovnali dátumu 2020-04-29
df2 #Výpis premennej df2
```

```python
may = pd.read_csv('tesla_may.csv', parse_dates=['datetime'], infer_datetime_format=True) #Načítanie druhého csv súboru pod názvom "may"
```

```python
may['date'] = may.datetime.dt.date #Vytvorenie stĺpca date- naplnenie zo stĺpca datetime (dátum)
may['time'] = may.datetime.dt.time #Vytvorenie stĺpca time- naplnenie zo stĺpca datetime (čas)
may['date'] = pd.to_datetime(may['date']) #Konvertovanie stĺpcu date na typ datetime
```

```python
df3=(may[may["date"]=='2020-05-01'][:160]) #Vytvorenie novej premennej df3 - k nej bolo priradených prvých 160 riadkov zo stĺpca date z premennej may, ktoré sa rovnali dátumu 2020-05-01
df3 #Výpis premennej df3
```
