```python
import numpy as np #Importovanie knižníc na spracovanie dát a súborov 
import pandas as pd
```

```python
df = pd.read_csv('tesla_april.csv', parse_dates=['datetime'], infer_datetime_format=True) # načítanie prvého csv súboru kotrý som pomenoval df
```

```python
df['date'] = df.datetime.dt.date #Vytvorenie stĺpca date- naplnenie zo stĺpca datetime (dátum)
df['time'] = df.datetime.dt.time #Vytvorenie stĺpca time- naplnenie zo stĺpca datetime (čas)
df #výpis súboru csv
```

```python
df['date'] = pd.to_datetime(df['date']) #Konvertovanie stĺpcu date na typ datetime
```

```python
df1=(df[df["date"]=='2020-03-30'][:160]) #Vytvorenie novej premennej df1 - k nej bolo priradených prvýchh 160 riadkov zo stĺpca date z premennej df, ktoré sa rovnali dátumu 2020-03-30
df1 #výpis premennej df1
```

```python
df2=(df[df["date"]=='2020-04-29'][:160]) #Vytvorenie novej premennej df2 - k nej bolo priradených prvých 160 riadkov zo stĺpca date z premennej df, ktoré sa rovnali dátumu 2020-04-29
df2 #Výpis premennej df2
```

```python
may = pd.read_csv('tesla_may.csv', parse_dates=['datetime'], infer_datetime_format=True) #Načítanie druhého csv súboru pod názvom "may"
```

```python
may['date'] = may.datetime.dt.date         #Vytvorenie stĺpca date- naplnenie zo stĺpca datetime (dátum)
may['time'] = may.datetime.dt.time         #Vytvorenie stĺpca time- naplnenie zo stĺpca datetime (čas)
may['date'] = pd.to_datetime(may['date']) #Konvertovanie stĺpcu date na typ datetime
```

```python
df3=(may[may["date"]=='2020-05-01'][:160]) #Vytvorenie novej premennej df3 - k nej bolo priradených prvých 160 riadkov zo stĺpca date z premennej may, ktoré sa rovnali dátumu 2020-05-01
df3 #Výpis premennej df3
```

```python
cistenie1 = df1[['open','high','low','close','date']] #Vytvorenie premennej cistenie1 - jej naplnenie premennou df1 potrebnými stĺpcami (open, high, low, close, date)
cistenie1 #Výpis premennej
```

```python
cistenie2 = df2[['open','high','low','close','date']]#Vytvorenie premennej cistenie2 - jej naplnenie premennou df2 potrebnými stĺpcami (open, high, low, close, date)
cistenie2 #Výpis premennej
```

```python
cistenie3 = df3[['open','high','low','close','date']]#Vytvorenie premennej cistenie3 - jej naplnenie premennou df3 potrebnými stĺpcami (open, high, low, close, date)
cistenie3 #Výpis premennej
```

```python
df_final = pd.concat((cistenie1,cistenie2,cistenie3)) #Vytvorenie premennej df_final - jej naplnenie predošlými troma premennými
df_final #Výpis obsahu premennej df_final
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.utils import to_categorical
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
X = df_final.iloc[:,[0,3]].values
print(df_final.shape)
y = [-1 if row[1] - row[0] < 0 else 1 for row in X]
y = y[1:]
y.append(0)

df_final[5] = y
```

```python
train,test = train_test_split(df_final, shuffle=True, train_size=0.8, test_size=0.2)
print(train.shape,test.shape)
```

```python
model = Sequential()
model.add(Dense(128,input_dim=4, activation='relu'))
model.add(Dense(256,input_dim=128, activation='relu'))
model.add(Dense(64, input_dim=256, activation='sigmoid'))
model.add(Dense(1))


model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
train_x = train.iloc[:,0:4].values
train_y = train.iloc[:, 5].values

print(train_x.shape, train_y.shape)

history = model.fit(train_x, train_y, epochs=100, batch_size=10,verbose=1)
plt.plot(history.history['accuracy'])
plt.show()
```

```python
print(history.history)

plt.plot(history.history['loss'])
plt.show()
```

```python
test_x, test_y = test.iloc[:,0:4].values, test.iloc[:,5].values
loss, accuracy = model.evaluate(test_x, test_y)
```
